**Passing the Extracted text to Gemma2B model**

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# Load Gemma2B
model_name = "google/gemma-2b"  # Replace with the exact Gemma2B model path
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
# Define the prompt
prompt_template = '''
    You are an intelligent Invoice understanding assistant. Use the context provided below to answer the following questions regarding the Invoice "
    and provide the answers in a JSON format:
    Context:{context}

    Questions:
    1. What is the Invoice Date?
    2. What is the Invoice Number?
    3. What is the Customer Name?
    4. What is the Total Amount in the Invoice?
   '''
prompt = prompt_template.format(context=cleaned_text)
prompt
```

```python
# Tokenize and generate response
inputs = tokenizer(prompt, return_tensors="pt", truncation=True, max_length=512)
outputs = model.generate(inputs["input_ids"], num_return_sequences=1,max_new_tokens=300)

# Decode the response
response = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(response)
```
